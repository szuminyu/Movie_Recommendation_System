Next, I am going to calculate the IMDB's Weighted Ratings formula for the top movie list.

Weighted Ratings =

       (v / v+m)*R + (m / v+m)*C
        v is the number of votes for the movie
        m is the minimum votes required to be listed in the chart
        R is the average rating of the movie
        C is the mean of voting score across the whole report

I will choose 95th percentile for m. Therefore, I have 2274 movies for the content-based analysis.

```python
movie['vote_average'] = pd.to_numeric(movie['vote_average'])
movie['vote_count'] = pd.to_numeric(movie['vote_count'])
C = np.mean(movie['vote_average'])
C
```

```python
m = movie['vote_count'].quantile(q = 0.95)
m
```

```python
#extract releasing year from the release date
movie['release_date']= pd.to_datetime(movie['release_date'], errors = 'coerce')
movie['release_date'].head(10)
movie['year'] = movie['release_date'].dt.year
movie['year']=movie['year'].fillna(0).astype('int32')
```

```python
#extract the data that is top 5 percentile

top = movie.loc[:, ['title', 'year', 'new_genres', 'vote_average', 'vote_count']]
top5 = top['vote_count'] >= m
top = top[top5]
top
```

```python
#Weighted Rankings Formula

def weighted_ranking(x):
    v = x['vote_count']
    R = x['vote_average']
    w = ((v/ (v+m))*R) + ((m / (v+m))*C)
    return w

top['weighted_ranking'] = weighted_ranking(top)
top = top.sort_values(by = ['weighted_ranking'], ascending = False)
top.head(10)
```

## Content-Based

I use NLP package - rake_nltk to analyze the content of movies. I choose cosine similarity to identify the similar movies. Based on the result from the cosine similarity matrix, I can generate movie recommendations. 

```python
from rake_nltk import Rake
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer


content = movie.loc[:, ['title', 'year', 'new_genres', 'overview']]

content.head(10)
```

```python
content = content[pd.notnull(content['overview'])]

#initialize the column
content['token'] = ""
```

Next, I am going to extract words from the overview of the movies using rake_nltk package.

```python
#extract the token from the corpus 

def get_phrase(text):
    r = Rake()
    r.extract_keywords_from_text(text)
    return r.get_ranked_phrases()

content['token'] = content.overview.apply(get_phrase)
```

```python
content.head(10)
```

```python
#apply cosine similarity

content['token'] = content['token'].astype(str).str.replace('\[|\]|\'', '')
content['token'].head(10)

#count matrix
count = CountVectorizer()
count_matrix = count.fit_transform(content['token'])

#cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)

print(cosine_sim)
```

The cosine similarity matrix shows that the diagonal is 1, which makes sense because each instance is identical to itself.

I can now build a system based on the score of cosine similarity.

```python
def recommendation(name, cosine_sim = cosine_sim):
    
    index = content[content['title'] == name].index[0]
    
    similar_movies_index = pd.Series(cosine_sim[index].ravel())
    
    #the first one is itself so start at the second one (=1)
    similar_movies_index = similar_movies_index.sort_values(ascending = False).iloc[1:11, ] 
    similar_movies = content.title[similar_movies_index.index]
        
    recommendation = [movies for movies in similar_movies]
    return recommendation 
```

## Collaborative Filters

```python
#import dataset
ratings = pd.read_csv('/Users/Sumin/Downloads/ratings_small.csv')
ratings = ratings.drop(columns = ['timestamp'])
ratings.head(10)
```

This dataset has 100,000 rows. Because of the limitation of my computer, I will only use 20000 instances from the dataset. As you can see, the dataset has no missing values. 

```python
ratings = ratings.sample(frac = 0.2)
print(ratings.info())
```

Now, I am going to use Surprise, which uses Sklearn API. It is a recommendation engine designed for sklearn. 
For the model, I use SVD, which is knowned as Singular Value Decomposition. It can calculate the ratings from users and recommend movies. I use RMSE to examine the accuracy.

```python
from surprise import Reader, Dataset, SVD, evaluate, accuracy
from surprise.model_selection import KFold
reader  = Reader(rating_scale = (1,5))
```

```python
data = Dataset.load_from_df(ratings, reader)

#define a CV iterator
kf = KFold(n_splits =3)

algo = SVD()

for trainset, testset in kf.split(data):
    algo.fit(trainset)
    predictions=algo.test(testset)
    
    accuracy.rmse(predictions, verbose = True)
```

RMSE looks good! I think SVD has done a great job!
